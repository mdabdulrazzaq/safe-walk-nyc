```python

import pandas as pd

# API endpoint
df_path = "/Users/abdulrazzaq/Documents/GitHub/safe-walk-nyc/NYPD_Complaint_Data_Current__Year_To_Date__20250126 (1).csv"
df = pd.read_csv(df_path)
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Data Types:\n", df.dtypes)
print("Summary Statistics:\n", df.describe())
print("Missing Values:\n", df.isnull().sum())
```

```python
df.head()
```

```python
df = df.drop(columns=['HADEVELOPT', 'HOUSING_PSA', 'TRANSIT_DISTRICT', 'New Georeferenced Column', 'Lat_Lon'])
```

```python
df = df.dropna(subset=['Latitude', 'Longitude' , 'CMPLNT_TO_DT','PD_CD'])
```

```python
df['CMPLNT_DATETIME'] = pd.to_datetime(
    df['CMPLNT_FR_DT'] + ' ' + df['CMPLNT_FR_TM'],
    format='%m/%d/%Y %H:%M:%S',  # Specify the correct format
    errors='coerce'  # Coerce invalid dates to NaT
)
```

```python
df = df.dropna(subset=[ 'CMPLNT_DATETIME'])
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
df = df[df['CMPLNT_DATETIME'] >= '2024-09-19']
```

```python
df = df[(df['Longitude'] >= -74.05) & (df['Longitude'] <= -73.75) &
        (df['Latitude'] >= 40.50) & (df['Latitude'] <= 40.90)]
```

```python
import json

# Create a GeoJSON structure
geojson = {
    "type": "FeatureCollection",
    "features": []
}

for _, row in df.iterrows():
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row['Longitude'], row['Latitude']]
        },
        "properties": {
            "crime_type": row['OFNS_DESC'],
            "date": row['CMPLNT_DATETIME'].strftime('%Y-%m-%d %H:%M:%S'),
            "borough": row['BORO_NM'],
            "offense_level": row['LAW_CAT_CD']
        }
    }
    geojson['features'].append(feature)

# Save GeoJSON to a file
with open('crime_data.geojson', 'w') as f:
    json.dump(geojson, f)
```
